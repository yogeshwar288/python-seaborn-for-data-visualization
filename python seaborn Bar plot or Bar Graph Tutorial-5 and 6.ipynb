# how to dra seaborn barplot/bar graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
#for bar plot
#sns.barplot()
#parameters
"""
Signature:
sns.barplot(
    x=None,
    y=None,
    hue=None,
    data=None,
    order=None,
    hue_order=None,
    estimator=<function mean at 0x01A14F10>,
    ci=95,
    n_boot=1000,
    units=None,
    seed=None,
    orient=None,
    color=None,
    palette=None,
    saturation=0.75,
    errcolor='.26',
    errwidth=None,
    capsize=None,
    dodge=True,
    ax=None,
    **kwargs,

"""
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,hue=tips_df.sex)
```

```python
#data parameter=
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

```python
#order
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order)
```

```python
#hue_order
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder)
```

```python
#estimator
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,estimator=np.max)
```

```python
# ci=95=configration interval
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,
            
            ci=20)
```

```python
#orient parametar
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,orient='h')
#agar x axis catagrical ho to work nhi karta
```

```python
#agar bar blot ko horizenatal  karna hoto
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(y='day',x='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,orient='h')
```

```python
#orient parametar
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='total_bill',y='size',hue='sex',data=tips_df,orient='h')
```

```python
#color
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,color='g')
```

```python
#palette
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,
            color='g',
           palette='Accent')
```

```python
#saturtion
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,
            color='g',
           saturation=0.4)
```

```python
#errcolor
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,
            color='g',
           saturation=0.4,
            errcolor='y')
```

```python
#errwidth
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,
            color='g',
           saturation=0.4,
           errwidth=5)
```

```python
#capsize
hueorder=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hueorder,
            color='g',
           saturation=0.4,
           errwidth=5,
           capsize=0.2)
```

```python
#dodge
#hueorder=['Female','Male']
#order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,
           dodge=False)
```

# Python seaborn part-6

```python
#keyword argument
kwargs={'alpha':0.8,'linewidth':4,'edgecolor':'g','linestyle':'--'}
sns.barplot(x='day',y='total_bill',data=tips_df,hue='sex',**kwargs)
```

```python
#background
sns.set(style='darkgrid')
kwargs={'alpha':0.8,'linewidth':4,'edgecolor':'g','linestyle':'--'}
ax=sns.barplot(x='day',y='total_bill',data=tips_df,hue='sex',**kwargs)
ax.set(title='days and total_bill relation',
      xlabel='days',
      ylabel='total_bill')
```

```python
plt.figure(figsize=(12,8))
sns.set(style='darkgrid')
kwargs={'alpha':0.8,'linewidth':4,'edgecolor':'g','linestyle':'--'}
sns.barplot(x='day',y='total_bill',data=tips_df,hue='sex',**kwargs)
plt.title("relation between total bill and days" ,fontsize=20)
plt.xlabel('days',fontsize=15)
plt.ylabel('total bill',fontsize=15)
plt.savefig('relation between total bill and days')
plt.show()
```
