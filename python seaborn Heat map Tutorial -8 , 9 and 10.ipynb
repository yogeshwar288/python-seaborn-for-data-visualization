# seaborn Heatmap part-2 

```python
plt.figure(figsize=(16,8))
annotkws={'fontsize':10,
          'fontstyle':'italic',
          'alpha':0.6,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w',
          'color':'k'}
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annotkws)
```

```python
#linewidth
plt.figure(figsize=(16,8))
annotkws={'fontsize':10,
          'fontstyle':'italic',
          'alpha':0.6,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w',
          'color':'k'}
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annotkws,linewidth=2)
```

```python
#linecolor
plt.figure(figsize=(16,8))
annotkws={'fontsize':10,
          'fontstyle':'italic',
          'alpha':0.6,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w',
          'color':'k'}
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annotkws,linewidth=2,linecolor='g')
```

```python
#linecolor
plt.figure(figsize=(16,8))
annotkws={'fontsize':10,
          'fontstyle':'italic',
          'alpha':0.6,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w',
          'color':'k'}
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annotkws,linewidth=2,linecolor='k')
```

```python
#if we won't labels xticslabel=false
plt.figure(figsize=(16,8))
annotkws={'fontsize':10,
          'fontstyle':'italic',
          'alpha':0.6,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w',
          'color':'k'}
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=annotkws,linewidth=2,linecolor='k',
           xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,xticklabels=False,yticklabels=False)
```

```python
#cbar_kws=if we want modification in color bar
plt.figure(figsize=(16,8))
cbar_kws={'orientation':"horizontal",
         'shrink':1,
         'extend':'min',
         'extendfrac':0.1,
         'ticks':np.arange(0,22),
         'drawedges':True}
sns.heatmap(globalwarming_df,cmap='coolwarm',annot=True,cbar_kws=cbar_kws)
```

```python
country_code=['usa','GBR','IND','CHN','RUS','AUS','FRA','DEU','CAN','BRA','ARG','PAK','NPL','BGD','JPN']
plt.figure(figsize=(16,8))
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap='coolwarm',annot=True,xticklabels=np.arange(0,15),yticklabels=country_code)
```

```python
plt.figure(figsize=(16,8))
ax=sns.heatmap(globalwarming_df,cmap='coolwarm',annot=True)
#ax.set(title='Who_is_responsible_for_global_warmin',
#      xlabel='CO2 emissions (metric tons per capita)',
 #     ylabel='Country Name')
plt.title('Who_is_responsible_for_global_warmin',fontsize=20)
plt.xlabel('CO2 emissions (metric tons per capita)',fontsize=15)
plt.ylabel('country name',fontsize=15,color='g')
sns.set(font_scale=1)
```

# seaborn heatmap part-3

# correlation

```python
#correlation
globalwarming_df.corr()
```

```python
#correlation with heatmap
#annot
#linewidth
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df.corr(),annot=True,linewidth=1)
```

```python
#correlation with heatmap
#annot
#linewidth
#ax.tick_params(parameters for labels)
plt.figure(figsize=(16,9))
ax=sns.heatmap(globalwarming_df.corr(),annot=True,linewidth=1)
ax.tick_params(size=10,labelsize=13,labelcolor='b')
plt.title('Who_is_responsible_for_global_warmin',fontsize=25)
plt.xlabel('CO2 emissions (metric tons per capita)',fontsize=15)
plt.ylabel('country name',fontsize=15)

plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
#datset me convert karte hai
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                      columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
#correlation with heatmap
#annot
#linewidth
#ax.tick_params(parameters for labels)
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(),annot=True,linewidth=1)
ax.tick_params(size=10,labelsize=25,labelcolor='b')
plt.title('breast_cancer_dataset',fontsize=25)
plt.show()
```
