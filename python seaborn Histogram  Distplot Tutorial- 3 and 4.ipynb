# Histogram Distplot Parametar

```python
#parameters,how we can use
'''
sns.distplot(
    a,
    bins=None,
    hist=True,
    kde=True,
    rug=False,
    fit=None,
    hist_kws=None,
    kde_kws=None,
    rug_kws=None,
    fit_kws=None,
    color=None,
    vertical=False,
    norm_hist=False,
    axlabel=None,
    label=None,
    ax=None,
)
'''
```

```python
#background change karna ho to sns.set(parametar) function a use karte hai

'''
sns.set(
    context='notebook',
    style='darkgrid',
    palette='deep',
    font='sans-serif',
    font_scale=1,
    color_codes=True,
    rc=None,
)
'''
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
sns.distplot(tips_df['total_bill'],bins=20,fit=norm,color='g',
             axlabel='Total Bill',label='total bill')
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.show
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill")
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",hist_kws={'color':'r','edgecolor':'b'})
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",hist_kws={'color':'y'})
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",hist_kws={'color':'k'})
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
histkws_parameter={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8}
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",
             hist_kws={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8},
             kde_kws={'color':'b','linewidth':2,'linestyle':'--','alpha':0.9})
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
histkws_parameter={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8}
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",
             hist_kws={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8},
             kde_kws={'color':'b','linewidth':2,'linestyle':'--','alpha':0.9},rug=True)
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
histkws_parameter={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8}
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",
             hist_kws={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8},
             kde_kws={'color':'b','linewidth':2,'linestyle':'--','alpha':0.9},
             rug=True,
             rug_kws={'color':'g','edgecolor':'r','linestyle':'--','alpha':0.8})
plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
histkws_parameter={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8}
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",
             hist_kws={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8},
             kde_kws={'color':'b','linewidth':2,'linestyle':'--','alpha':0.9},
             rug=True,
             rug_kws={'color':'g','edgecolor':'r','linestyle':'--','alpha':0.8},
             kde=False,
             fit=norm,
             fit_kws={'color':'y','linewidth':2,'linestyle':'--','alpha':0.9})

plt.title('total bill histogram',fontsize=20)#for title
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
histkws_parameter={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8}
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",
             hist_kws={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8},
             kde_kws={'color':'b','linewidth':2,'linestyle':'--','alpha':0.9},
             rug=True,
             rug_kws={'color':'g','edgecolor':'r','linestyle':'--','alpha':0.8},
             kde=False,
             fit=norm,
             fit_kws={'color':'y','linewidth':2,'linestyle':'--','alpha':0.9})

plt.title('total bill histogram',fontsize=20)#for title
plt.xlabel('totalbill',fontsize=15)
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(12,7))
#for total_bill
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill")
#fortips
sns.distplot(tips_df['tip'],bins=bins,color='y',label="tips")
#for size
sns.distplot(tips_df['size'],bins=bins,color='b',label="size")
plt.title('total bill histogram',fontsize=20)#for title
plt.xlabel('totalbill',fontsize=15)
plt.legend()#for label
plt.xticks()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set(style='darkgrid')
plt.figure(figsize=(18'9))
histkws_parameter={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8}
sns.distplot(tips_df['total_bill'],bins=bins,color='g',label="totalbill",
             hist_kws={'color':'g','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':0.8},
             kde_kws={'color':'b','linewidth':2,'linestyle':'--','alpha':0.9},
             rug=True,
             rug_kws={'color':'g','edgecolor':'r','linestyle':'--','alpha':0.8},
             kde=False,
             fit=norm,
             fit_kws={'color':'y','linewidth':2,'linestyle':'--','alpha':0.9})
#fortips
sns.distplot(tips_df['tip'],bins=bins,color='y',label="tips")
#for size
sns.distplot(tips_df['size'],bins=bins,color='b',label="size")


plt.title('total bill histogram',fontsize=20)#for title
plt.xlabel('totalbill',fontsize=15)
plt.legend()#for label
plt.xticks()
plt.show()
```
