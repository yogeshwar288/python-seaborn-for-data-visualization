# How to draw seaborn Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df
```

```python
#we will use for scatter sns.scatterplot() function
sns.scatterplot()
'''
sns.scatterplot(
    x=None,
    y=None,
    hue=None,
    style=None,
    size=None,
    data=None,
    palette=None,
    hue_order=None,
    hue_norm=None,
    sizes=None,
    size_order=None,
    size_norm=None,
    markers=True,
    style_order=None,
    x_bins=None,
    y_bins=None,
    units=None,
    estimator=None,
    ci=95,
    n_boot=1000,
    alpha='auto',
    x_jitter=None,
    y_jitter=None,
    legend='brief',
    ax=None,
    **kwargs,

'''
```

```python
titanic_df
```

```python
#data ko scatter karte hai x=age y=fare
sns.scatterplot(x='age',y='fare',data=titanic_df)
```

```python
#hue
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex')
```

```python
#style parameter me 'who' ka use karege jisse child, men women devide ho jayege
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who')
```

```python
#size=male our female ke scatter ka size bd jayega our child ka size chota ho jayega
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who')
```

```python
#sizes=agar scatter point ke sizes ko bdana ho to sizes=(min,max)
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who',sizes=(20,200))
```

```python
#for big figure
plt.figure(figsize=(14,8))
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who',sizes=(20,200))
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x='who',y='fare',data=titanic_df,hue='alive',style='alive',size='who',sizes=(20,200))
```

```python
#palette
plt.figure(figsize=(14,8))
sns.scatterplot(x='who',y='fare',data=titanic_df,hue='alive',style='alive',size='who',sizes=(20,200),
               palette='hot',alpha=0.6)
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x='age',y='fare',data=titanic_df)
sns.lineplot(x='age',y='fare',data=titanic_df)
sns.barplot(x='age',y='fare',data=titanic_df)
```
