# seaborn pair plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
#pairplot
ax=sns.pairplot(cancer_df,hue='target')
ax
```

```python
#vars function

sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'])
```

```python
#hue function
sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],hue='target')
```

```python
#palette
sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],hue='target', palette='Accent')
```

```python
#palette
sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],hue='target', palette='Dark2')
```

```python
#x_vars=[],,yvars[]
sns.pairplot(cancer_df,x_vars=['mean radius', 'mean texture'],y_vars=['mean radius'],hue='target', palette='Dark2')
```

```python
#kind

sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],
             hue='target', palette='Dark2',kind='reg')
```

```python
#diag_kind

sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],
             hue='target', palette='Dark2',diag_kind='hist')
```

```python
#marker

sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],
             hue='target', palette='Dark2',diag_kind='hist',markers=['*','>'])
```

```python
#height

sns.pairplot(cancer_df,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness'],
             hue='target', palette='Dark2',diag_kind='hist',markers=['*','>'],height=20)
```
